```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
import seaborn
```

```python
zoren = yf.Ticker("ZOREN.IS")
```

```python
zoren = zoren.history(period="max")
```

```python
zoren
```

```python
zoren.index
```

```python
#we cleaned and visualized the data

zoren.plot.line(y="Close", use_index=True)
```

```python
del zoren["Dividends"]
del zoren["Stock Splits"]
```

```python
 zoren["Tomorrow"] = zoren["Close"].shift(-1)
```

```python
zoren
```

```python
# we set up our machine learning target

zoren ["Target"] = (zoren["Tomorrow"] > zoren["Close"]).astype(int)
```

```python
zoren
```

```python
zoren = zoren.loc["2002-12-19 00:00:00+03:00":].copy()
```

```python
zoren
```

```python
# we trained our initial model

from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators=100, min_samples_split=100, random_state=1)

train = zoren.iloc[:-100]
test = zoren.iloc[-100:]

predictors= ["Close", "Volume" , "High", "Low"]
model.fit(train[predictors], train["Target"])
```

```python
from sklearn.metrics import precision_score

preds = model.predict(test[predictors])
```

```python
preds = pd.Series(preds, index=test.index)
```

```python
precision_score(test["Target"], preds)
```

```python
combined = pd.concat([test["Target"], preds], axis=1)
```

```python
combined.plot()
```

```python
def predict(train, test, predictors, model):
    model.fit(train[predictors], train["Target"])
    preds = model.predict(test[predictors])
    preds = pd.Series(preds, index=test.index, name="Predictions")
    combined = pd.concat([test["Target"], preds], axis=1)
    return combined
```

```python
# we evaulted error and created a way to backtest


def backtest(data, model, predictors, start=2500, step=250):    
    all_predictions = []
    
    for i in range(start, data.shape[0], step):
        train = data.iloc[0:i].copy()
        test = data.iloc[i:(i+step)].copy()
        predictions = predict(train, test, predictors, model)
        all_predictions.append(predictions)
    return pd.concat(all_predictions)    
```

```python
predictions = backtest(zoren, model, predictors)
```

```python
# we predicted market would go down approx 2643 days and market would go up 144 days


predictions["Predictions"].value_counts()
```

```python
# we accomplished that when we say market would go up , we corrected right %37.5 of the time

precision_score(predictions["Target"], predictions["Predictions"])
```

```python
# we can say the stock market went up %40 of the time and our model predicted that it would go up %37.5 of the time.

predictions["Target"].value_counts() / predictions.shape[0]
```

```python
# we improved our model with extra predictor columns


horizons = [2,5,60,250,1000]
new_predictors = []

for horizon in horizons:
    rolling_averages = zoren.rolling(horizon).mean()
    
    ratio_column = f"Close_Ratio_{horizon}"
    zoren[ratio_column] = zoren["Close"] / rolling_averages["Close"]
    
    trend_column = f"Trend_{horizon}"
    zoren[trend_column] = zoren.shift(1).rolling(horizon).sum()["Target"] #sum of the days that stock price actually went up
    
    new_predictors += [ratio_column, trend_column]
```

```python
zoren = zoren.dropna()
zoren
```

```python
model = RandomForestClassifier(n_estimators=200, min_samples_split=50, random_state=1)
```

```python
def predict(train, test, predictors, model):
    model.fit(train[predictors], train["Target"])
    preds = model.predict_proba(test[predictors])[:,1]
    preds[preds >= .6] = 1
    preds[preds < .6] = 0
    preds = pd.Series(preds, index=test.index, name="Predictions")
    combined = pd.concat([test["Target"], preds], axis=1)
    return combined
```

```python
predictions = backtest(zoren, model, new_predictors)
```

```python
predictions["Predictions"].value_counts()
```

```python
#after adding predictors and improving our model , we can see our precision score went up and model got more confident

precision_score(predictions["Target"], predictions["Predictions"])
```
